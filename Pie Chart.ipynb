```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
exp_values=[1400,600,300,410,250]
exp_labels=["Home Rent","Food","Phone/Internet Bill","Car","Other Utilities "]
```

```python
plt.pie(exp_values,labels=exp_labels)
```

```python
plt.axis('equal')
plt.pie(exp_values,labels=exp_labels)
plt.show() #this line is to avoid the texts around the pie chart
```

```python
#to increase the size of the pie chart, use the radius argument
plt.pie(exp_values,labels=exp_labels,radius=2)
plt.show()
```

```python
#to get the percentified explanation
plt.axis=0
plt.pie(exp_values,labels=exp_labels,autopct='%0.2f%%')#.2 indicates two digits after decimal.For no digits after decimal,.0
plt.show()
```

```python
#to add shadow
plt.axis=0
plt.pie(exp_values,labels=exp_labels,autopct='%0.2f%%',shadow=True)
plt.show()
```

```python
#to explode the pie chart

plt.pie(exp_values,labels=exp_labels,autopct='%0.2f%%',explode=[0,0.2,0,0,0])#0.2 is for the pie that needs to be exploded
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_values,labels=exp_labels,autopct='%0.2f%%',explode=[0,0.2,0,0,0],startangle=45)
plt.show()
```
