```python
company=["GOOGL","AMZN","MSFT","FB"]
revenue=[90,136,39,27]
```

```python
import numpy as np
xpos=np.arange(len(company))
#coz bar plot won't honour string values
```

```python
plt.bar(xpos,revenue)
```

```python
plt.xticks(xpos,company)
plt.xlabel("COMPANY")
plt.ylabel("REVENUE")
plt.title("US Tech Stocks")
plt.bar(xpos,revenue,color='orange')
```

```python
profit=[40,2,34,12]
```

```python
plt.xticks(xpos,company)
plt.xlabel("COMPANY")
plt.ylabel("REVENUE")
plt.title("US Tech Stocks")
plt.bar(xpos,revenue)
plt.bar(xpos,profit)
```

```python
plt.xticks(xpos,company)
plt.xlabel("COMPANY")
plt.ylabel("REVENUE")
plt.title("US Tech Stocks")
plt.bar(xpos,revenue,label="Revenue")
plt.bar(xpos,profit,label="Profit")
plt.legend()
```

```python
plt.xticks(xpos,company)
plt.xlabel("COMPANY")
plt.ylabel("REVENUE")
plt.title("US Tech Stocks")
plt.bar(xpos-0.2,revenue,label="Revenue")
plt.bar(xpos+0.3,profit,label="Profit")
plt.legend()
```

```python
plt.xticks(xpos,company)
plt.xlabel("COMPANY")
plt.ylabel("REVENUE")
plt.title("US Tech Stocks")
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="Profit")
plt.legend()
```

```python
plt.yticks(xpos,company)
plt.xlabel("COMPANY")
plt.ylabel("REVENUE")
plt.title("US Tech Stocks")
plt.barh(xpos,revenue,label="Revenue")
plt.barh(xpos,profit,label="Profit")
plt.legend()
```

```python
blood_sugar=[113,85,90,150,149,88,93,115,135,80,77,82,129]
```

```python
plt.hist(blood_sugar)
```

```python
plt.hist(blood_sugar,bins=3)
```

```python
plt.hist(blood_sugar,bins=3,rwidth=0.5)
```

```python
plt.hist(blood_sugar,bins=[80,100,125,200],rwidth=0.5)
#customizing the bin parameter explicitly
```

```python
plt.hist(blood_sugar,bins=[80,100,125,200],rwidth=0.5,histtype='step')
```

```python
blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women=[67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['green','orange'])
```

```python
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['green','orange'],label=['men','women'])
plt.legend()
```

```python
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.95,color=['green','orange'],label=['men','women'],orientation='horizontal')
plt.legend()
```
